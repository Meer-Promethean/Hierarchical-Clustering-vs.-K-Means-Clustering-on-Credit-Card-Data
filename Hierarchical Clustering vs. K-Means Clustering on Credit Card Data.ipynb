```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load the dataset
data = pd.read_csv('CC GENERAL.csv')

# Step 1: DATA PREPROCESSING :: 
#Inspect Dataset Structure
print("Dataset Shape:", data.shape)
print("Column Names:", data.columns.tolist())
print("Missing Values:\n", data.isnull().sum())


```

```python
#So missing values in CREDIT_LIMIT and in MINIMUM_PAYMENTS
# Handle Missing Values
# Replace missing values with the median for numerical columns
data['CREDIT_LIMIT'].fillna(data['CREDIT_LIMIT'].median(), inplace=True)
data['MINIMUM_PAYMENTS'].fillna(data['MINIMUM_PAYMENTS'].median(), inplace=True)

```

```python
print("Missing Values:\n", data.isnull().sum())
```

```python
# Drop Non-Numerical Columns
# 'CUST_ID' is non-numerical and irrelevant for clustering, so we drop it.
data_numeric = data.drop(columns=['CUST_ID'])

# Step 4: Normalize/Scale Numerical Features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_numeric)

# Create a DataFrame with scaled values for reference
data_scaled_df = pd.DataFrame(data_scaled, columns=data_numeric.columns)

# Verify Preprocessed Data
print("Missing Values After Processing:\n", data_scaled_df.isnull().sum())
print("Sample Scaled Data:\n", data_scaled_df.head())
```

```python
# Step 2 : EDA
import seaborn as sns
import matplotlib.pyplot as plt
# Exclude non-numeric columns for correlation analysis
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Generate Summary Statistics
print("Summary Statistics:\n", numeric_data.describe())

# Visualize Distributions of Key Features
key_features = ['BALANCE', 'PURCHASES', 'CREDIT_LIMIT', 'PAYMENTS', 'MINIMUM_PAYMENTS']
for feature in key_features:
    if feature in data.columns:  # Ensure the column exists
        plt.figure(figsize=(6, 4))
        sns.histplot(data[feature], kde=True, bins=30, color='blue')
        plt.title(f"Distribution of {feature}")
        plt.xlabel(feature)
        plt.ylabel("Frequency")
        plt.show()

# Correlation Analysis
correlation_matrix = numeric_data.corr()  # Compute correlation for numeric data only
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Matrix")
plt.show()

# Pairwise Scatter Plots
sns.pairplot(numeric_data[key_features], diag_kind='kde', corner=True)
plt.suptitle("Pairwise Scatter Plots", y=1.02)
plt.show()
```

```python
# Step 3: Hierarchical Clustering

from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Perform Hierarchical Clustering
# Compute the linkage matrix using Ward's method
linked = linkage(data_scaled, method='ward')

# Compute and Visualize the Dendrogram
plt.figure(figsize=(12, 8))
dendrogram(linked, truncate_mode='level', p=5)  # Show top 5 levels of the hierarchy
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data Points or Clusters")
plt.ylabel("Euclidean Distance")
plt.show()

# Choose Optimal Number of Clusters
# For example, cut the dendrogram at 4 clusters
optimal_clusters = 4
cluster_labels = fcluster(linked, t=optimal_clusters, criterion='maxclust')

# Assign cluster labels to the data
data['Cluster_Labels'] = cluster_labels

#Inspect the cluster assignment
print("Cluster Counts:\n", data['Cluster_Labels'].value_counts())
print("Sample Data with Cluster Labels:\n", data[['Cluster_Labels']].head())
```

```python
#Step 4 K-Means Clustering

from sklearn.cluster import KMeans
# Train the K-Means Model
# Define the number of clusters
optimal_clusters = 4
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
data['Cluster_Labels_KMeans'] = kmeans.fit_predict(data_scaled)

# Inspect Cluster Assignments
print("Cluster Counts:\n", data['Cluster_Labels_KMeans'].value_counts())
print("Sample Data with Cluster Labels:\n", data[['Cluster_Labels_KMeans']].head())

# Visualize the Clusters
# Use the first two principal components (for simplicity, use the first two scaled features)
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=data_scaled[:, 0],
    y=data_scaled[:, 1],
    hue=data['Cluster_Labels_KMeans'],
    palette='viridis',
    s=50
)
plt.title("K-Means Clustering Visualization")
plt.xlabel("Feature 1 (Scaled)")
plt.ylabel("Feature 2 (Scaled)")
plt.legend(title="Cluster", loc='upper right')
plt.show()
```

```python

# Normalize the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_data = pd.DataFrame(scaler.fit_transform(data.select_dtypes(include=['float64', 'int64'])),
                               columns=data.select_dtypes(include=['float64', 'int64']).columns)
```

```python
from scipy.cluster.hierarchy import linkage, fcluster
from sklearn.cluster import KMeans

# Hierarchical clustering
linkage_matrix = linkage(normalized_data, method='ward')
normalized_data['Hierarchical_Cluster'] = fcluster(linkage_matrix, t=7, criterion='distance')

# K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
normalized_data['KMeans_Cluster'] = kmeans.fit_predict(normalized_data)
```

```python
hierarchical_cluster_counts = normalized_data['Hierarchical_Cluster'].value_counts()
print("Hierarchical Clustering Counts:\n", hierarchical_cluster_counts)

kmeans_cluster_counts = normalized_data['KMeans_Cluster'].value_counts()
print("K-Means Clustering Counts:\n", kmeans_cluster_counts)
```

```python
hierarchical_centroids = normalized_data.groupby('Hierarchical_Cluster').mean()
print("Hierarchical Cluster Centroids:\n", hierarchical_centroids)
```

```python
kmeans_centroids = pd.DataFrame(kmeans.cluster_centers_, columns=normalized_data.columns[:-2])
print("K-Means Cluster Centroids:\n", kmeans_centroids)
```

```python
# Cross-tabulation of cluster memberships
comparison_table = pd.crosstab(normalized_data['Hierarchical_Cluster'], normalized_data['KMeans_Cluster'])
print("Cluster Comparison Table:\n", comparison_table)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualizing two key features colored by clusters from both methods
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=normalized_data['BALANCE'], 
    y=normalized_data['PURCHASES'], 
    hue=normalized_data['Hierarchical_Cluster'], 
    style=normalized_data['KMeans_Cluster'], 
    palette='Set1'
)
plt.title("Comparison of Hierarchical and K-Means Clusters")
plt.show()
```
